```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example data (replace with actual DataFrame)
dominos_data = pd.DataFrame({
    'order_date': ['2024-11-01', '2024-11-02', '2024-11-03', '2024-11-04', '2024-11-05'],
    'total_price': [1500, 2000, 1800, 1600, 1900]
})

# Prepare data
dominos_data['order_date'] = pd.to_datetime(dominos_data['order_date'])
dominos_data['weekday'] = dominos_data['order_date'].dt.day_name()
dominos_data['weekday'] = pd.Categorical(dominos_data['weekday'], categories=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], ordered=True)

# Group by weekday and sum the total revenue
weekday_revenue = dominos_data.groupby('weekday')['total_price'].sum()

# Plot with Seaborn style
sns.set_style('whitegrid')
plt.figure(figsize=(12, 6))
weekday_revenue.plot(kind='bar', color=sns.color_palette('coolwarm', len(weekday_revenue)), edgecolor='black', width=0.8)

# Customize plot
plt.title('Total Revenue by Weekday', fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel('Day of the Week', fontsize=14, fontweight='bold', color='darkblue')
plt.ylabel('Total Revenue', fontsize=14, fontweight='bold', color='darkblue')
plt.xticks(rotation=45, fontsize=12, color='darkblue')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()




```

```python
# Example data (replace this with your actual DataFrame)
data = {
    'order_date': ['2024-11-01', '2024-11-02', '2024-11-03', '2024-12-01', '2024-12-02'],
    'total_price': [1500, 2000, 1800, 1600, 1900],
}
dominos_data = pd.DataFrame(data)

# Convert order_date to datetime if it isn't already
dominos_data['order_date'] = pd.to_datetime(dominos_data['order_date'])

# Extract month and year from order_date
dominos_data['month'] = dominos_data['order_date'].dt.to_period('M')

# Group by month and sum the total revenue
monthly_revenue = dominos_data.groupby('month')['total_price'].sum()

# Plotting the line chart
plt.figure(figsize=(10, 6))
monthly_revenue.plot(kind='line', marker='o', color='orange')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.show()
```
